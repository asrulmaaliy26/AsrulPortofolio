# **Bab 3: Logika dan Algoritma Dasar**

---

## **Pertemuan 1: Melatih Logika Berpikir Komputasional**

### **A. Pendahuluan (10 menit)**

**Guru membuka:**

> “Pernah nggak kamu memberi instruksi ke teman, tapi dia salah mengerti? Ternyata, menyusun instruksi itu ada seninya! Nah, di dunia komputer, kita menyebutnya dengan **logika dan algoritma**.”

**Tujuan disampaikan kepada siswa:**

* Memahami pentingnya logika dalam menyusun perintah.
* Melatih kemampuan berpikir sistematis dan runut.

---

### **B. Isi Materi (60 menit)**

#### **1. Pengenalan Konsep Logika (15 menit)**

**Penjelasan guru:**

* Logika komputasional adalah dasar berpikir sistematis seperti komputer.
* Komputer hanya menjalankan perintah yang **tepat dan jelas**.

**Contoh:**

> “Ambil segelas air” – terdengar mudah, tapi bisa kacau jika instruksi tidak jelas (misalnya, tidak tahu letak gelas, tidak tahu cara menuang).

---

#### **2. Aktivitas Teka-Teki Logika (20 menit)**

**Contoh aktivitas:**

* Permainan “logika 3 kotak”: logika pernyataan benar/salah.
* Kuis logika:

  * "Semua kucing adalah hewan. Beberapa hewan bisa berenang. Apakah semua kucing bisa berenang?" (→ **Tidak**)
  * “Jika lampu menyala, maka ruangan terang. Ruangan gelap. Apakah lampu menyala?” (→ **Tidak pasti**)

**Tujuan:** Melatih logika deduktif dan sebab-akibat.

---

#### **3. Aktivitas “Jadilah Robot” (25 menit)**

**Tugas berpasangan:**

* Satu siswa menjadi “robot”, satu menjadi “programmer”.
* Tujuan: membuat robot menyusun tumpukan balok atau berjalan ke tujuan (bisa di lantai kelas).
* Programmer hanya boleh memberi instruksi satu per satu (misal: "maju 1 langkah", "belok kiri").

**Refleksi:**

> “Apa yang terjadi jika instruksi salah atau tidak lengkap?” → robot bingung atau salah gerak (simulasi logika komputer).

---

### **C. Penutup (10 menit)**

**Guru menyimpulkan:**

> “Komputer tidak bisa menebak. Semua harus diberikan secara jelas dan logis. Makanya, logika itu penting dalam informatika.”

**Tugas rumah:**

> “Tulis 3 kegiatan sehari-harimu dan pecah jadi langkah-langkah logis (misalnya: cara membuat mie instan).”

---

## **Pertemuan 2: Mengenal Algoritma, Flowchart, dan Pseudocode**

### **A. Pendahuluan (5–10 menit)**

**Guru membuka:**

> “Kalian sudah tahu pentingnya logika. Hari ini, kita akan belajar cara menuliskan logika itu secara terstruktur: **dalam bentuk algoritma, diagram alur, dan pseudocode.**”

**Tujuan:**

* Siswa memahami bentuk representasi algoritma.
* Siswa mampu menyusun flowchart dan pseudocode dari aktivitas sehari-hari.

---

### **B. Isi Materi (60 menit)**

#### **1. Penjelasan Algoritma (10 menit)**

**Penjelasan guru:**

* Algoritma = langkah-langkah terurut untuk menyelesaikan masalah.
* Contoh sederhana: Algoritma membuat teh.

---

#### **2. Pengenalan Flowchart (10 menit)**

**Simbol dasar:**

* **Oval**: Mulai/Selesai
* **Persegi panjang**: Proses/Kegiatan
* **Belah ketupat**: Keputusan/percabangan
* **Panah**: Alur

**Contoh ditampilkan:** Flowchart membuat mie instan.

---

#### **3. Pengenalan Pseudocode (10 menit)**

**Definisi:**

* Bentuk penulisan algoritma mirip kode, tapi dalam bahasa manusia.

**Contoh:**

```
Mulai
Isi teko dengan air
Rebus air
Masukkan teh ke dalam gelas
Tuang air panas ke gelas
Aduk
Selesai
```

---

#### **4. Aktivitas Kelompok: Buat Flowchart & Pseudocode "Membuat Teh" (30 menit)**

**Langkah:**

* Tiap kelompok (3–4 orang) membuat:

  * Flowchart "cara membuat teh"
  * Pseudocode-nya
* Presentasi cepat (1 menit per kelompok)
* Guru dan teman menilai: Apakah langkahnya logis dan urut?

---

### **C. Penutup (10 menit)**

**Guru menyimpulkan:**

> “Algoritma, flowchart, dan pseudocode adalah cara untuk menyampaikan logika secara runtut dan jelas. Semakin jelas algoritma kita, semakin mudah komputer memahaminya.”

**Refleksi akhir:**

> “Jika komputer hanya bisa menjalankan instruksi, siapa yang harus berpikir?” → Jawaban: **manusia (kita)**

# **Bab 4: HTML – Struktur Halaman Web**

---

## **Pertemuan 1: Mengenal Struktur Dasar HTML**

### **A. Pendahuluan (10 menit)**

**Guru membuka:**

> “Setiap halaman web yang kalian buka—entah itu YouTube, Wikipedia, atau web sekolah—dibangun dengan sesuatu yang bernama HTML. Hari ini, kita akan belajar bagaimana struktur sebuah halaman web itu dibuat!”

**Tujuan disampaikan kepada siswa:**

* Mengenal fungsi HTML sebagai kerangka halaman web.
* Mengetahui struktur dasar HTML dan tag-tag penting.

---

### **B. Isi Materi (60 menit)**

#### **1. Apa Itu HTML? (10 menit)**

**Penjelasan guru:**

* HTML = **HyperText Markup Language** → digunakan untuk **membuat struktur halaman web**.
* Bukan bahasa pemrograman, tapi bahasa markup.
* HTML menyusun konten seperti paragraf, gambar, link, judul.

**Analogi sederhana:**

> HTML itu seperti “rangka” atau “tulang” dari tubuh web, CSS itu “pakaiannya”, dan JavaScript itu “otaknya”.

---

#### **2. Struktur Dasar HTML (15 menit)**

**Ditulis di papan atau slide:**

```html
<!DOCTYPE html>
<html>
  <head>
    <title>Judul Halaman</title>
  </head>
  <body>
    <h1>Halo Dunia!</h1>
    <p>Ini adalah paragraf pertama saya.</p>
  </body>
</html>
```

**Penjelasan bagian-bagian:**

* `<!DOCTYPE html>` → mendefinisikan dokumen HTML5
* `<html>` → pembungkus seluruh dokumen
* `<head>` → informasi, judul, meta
* `<body>` → isi halaman: teks, gambar, tombol, dll.

---

#### **3. Demo Langsung HTML (15 menit)**

**Guru membuka editor online seperti [replit.com](https://replit.com), [jsfiddle.net](https://jsfiddle.net), atau [codepen.io](https://codepen.io)**

* Mengetik kode HTML di depan kelas.
* Tunjukkan hasil langsung di browser.

**Tujuan:** Siswa memahami hubungan antara kode dan tampilan visual.

---

#### **4. Aktivitas Mandiri: Coba Tulis HTML (20 menit)**

**Siswa menulis kode HTML sederhana:**

```html
<!DOCTYPE html>
<html>
  <head>
    <title>Halaman Pertamaku</title>
  </head>
  <body>
    <h1>Halo, saya [Nama]</h1>
    <p>Saya sedang belajar HTML di kelas informatika.</p>
  </body>
</html>
```

**Langkah:**

* Bisa dikerjakan di laptop atau HP via editor online.
* Siswa menyimpan atau screenshot hasilnya.

# **Bab 5: CSS – Desain Tampilan**

---

## **Pertemuan 1: Mengenal CSS dan Peranannya dalam Web**

### **A. Pendahuluan (10 menit)**

**Guru membuka:**

> “Coba bandingkan halaman web Google awal (putih polos) dengan halaman YouTube atau Instagram. Apa bedanya? Yap—warnanya, bentuk tombolnya, ukuran teksnya—semua itu dibuat dengan CSS!”

**Tujuan disampaikan kepada siswa:**

* Mengenal fungsi CSS dalam desain halaman web.
* Melihat perbedaan tampilan halaman dengan dan tanpa CSS.

---

### **B. Isi Materi (60 menit)**

#### **1. Apa Itu CSS? (10 menit)**

**Penjelasan guru:**

* CSS = *Cascading Style Sheets*
* Digunakan untuk **mengatur tampilan/visual halaman web**.
* Memisahkan konten (HTML) dengan gaya (CSS), agar web lebih rapi dan fleksibel.

**Analogi:**

> “Kalau HTML adalah kerangka rumah, CSS adalah cat tembok, dekorasi, dan perabotannya.”

---

#### **2. Perbandingan Web Tanpa vs. Dengan CSS (10 menit)**

**Tampilkan demo dua halaman:**

* HTML polos (tanpa CSS)
* HTML yang sama tapi sudah diberi CSS (warna, ukuran, layout)

**Diskusi cepat:**

> “Menurutmu, kenapa kita butuh CSS? Apa pengaruhnya terhadap pengalaman pengguna?”

---

#### **3. Sintaks Dasar CSS (15 menit)**

**Contoh kode:**

```css
h1 {
  color: blue;
  font-size: 24px;
}
```

**Penjelasan:**

* **Selector**: elemen yang ingin diberi gaya (`h1`)
* **Property**: jenis pengaturan (`color`, `font-size`)
* **Value**: nilainya (`blue`, `24px`)

**3 cara menyisipkan CSS:**

* Inline (di dalam tag HTML)
* Internal (di `<style>` dalam `<head>`)
* Eksternal (file `.css` terpisah)

---

#### **4. Aktivitas Siswa: Coba Tambah CSS ke HTML (25 menit)**

**Langkah:**

* Siswa membuka halaman HTML biodata yang sudah dibuat di Bab 4.
* Tambahkan CSS internal ke `<head>`:

```html
<style>
  body {
    background-color: #f0f0f0;
    font-family: Arial;
  }
  h1 {
    color: darkblue;
  }
  p {
    color: #333;
  }
</style>
```

**Output:** Halaman biodata terlihat lebih menarik secara visual.

---

**Guru menyimpulkan:**

# **Bab 6: JavaScript Dasar – Membuat Web Jadi Hidup**

---

## **Pertemuan 1: Pengenalan JavaScript dan Interaksi Dasar**

### **A. Pendahuluan (10 menit)**

**Guru membuka dengan:**

> “Kalau HTML adalah rangka dan CSS adalah gaya, maka JavaScript adalah otak dari halaman web. Dengan JavaScript, kita bisa membuat halaman web jadi interaktif—seperti tombol yang bisa diklik, muncul alert, sampai kalkulator!”

**Tujuan disampaikan kepada siswa:**

* Mengetahui fungsi JavaScript dalam membuat web interaktif.
* Mencoba sintaks dasar seperti `alert`, `prompt`, dan `variabel`.

---

### **B. Isi Materi (60 menit)**

#### **1. Apa Itu JavaScript? (10 menit)**

**Penjelasan guru:**

* JavaScript = bahasa pemrograman web.
* Berjalan di browser, bukan server.
* Bisa memanipulasi isi HTML, merespons aksi pengguna (klik, ketik, dll).

**Contoh tampilan nyata:**

* Alert popup
* Form input
* Game atau tombol warna berubah saat diklik

---

#### **2. Sintaks Dasar JavaScript (15 menit)**

**Penjelasan di papan atau editor online:**

```html
<script>
  alert("Halo, selamat datang!");
</script>
```

**Bahas:**

* `alert()` → menampilkan pesan ke pengguna
* `prompt()` → meminta input pengguna
* Variabel → menyimpan data

```javascript
let nama = prompt("Siapa namamu?");
alert("Halo " + nama + "!");
```

---

#### **3. Aktivitas Mandiri: “Halo Pengunjung” (35 menit)**

**Tugas siswa:**

* Buka HTML sederhana.
* Tambahkan kode JavaScript berikut:

```html
<script>
  let nama = prompt("Masukkan nama kamu:");
  alert("Halo " + nama + ", selamat belajar JavaScript!");
</script>
```

**Langkah:**

* Jalankan dan coba beberapa kali.
* Ganti teks alert sesuai kreativitas masing-masing.

---

### **C. Penutup (10 menit)**

**Guru menyimpulkan:**

> “Dengan JavaScript, kita bisa berbicara dengan pengguna melalui kotak pesan, menangkap input mereka, dan meresponsnya. Ini baru permulaan!”

**Tugas rumah:**

> “Tulis 3 ide web sederhana yang kamu ingin buat yang memiliki interaksi dengan pengguna.”

## **Pertemuan 2: Percabangan dan Tombol Interaktif**

### **A. Pendahuluan (5–10 menit)**

**Guru membuka dengan:**

> “Bayangkan kamu membuat kuis: jika jawabannya benar, tampilkan ‘selamat’, kalau salah, tampilkan ‘coba lagi’. Untuk itu, kita butuh **percabangan** dalam JavaScript.”

**Tujuan:**

* Memahami konsep `if`, `else`, dan `event` (tombol).
* Membuat interaksi berdasarkan kondisi logika.

---

### **B. Isi Materi (60 menit)**

#### **1. Percabangan If/Else (15 menit)**

**Contoh kode:**

```javascript
let nilai = prompt("Berapa nilai ujian kamu?");
if (nilai >= 75) {
  alert("Selamat, kamu lulus!");
} else {
  alert("Maaf, kamu belum lulus.");
}
```

**Penjelasan:**

* Gunakan operator `>`, `<`, `>=`, `==` untuk membuat logika.
* `if` = jika kondisi benar, lakukan.
* `else` = jika tidak, lakukan alternatifnya.

---

#### **2. Menambahkan Tombol Interaktif (20 menit)**

**Contoh HTML + JavaScript:**

```html
<button onclick="tampilkanPesan()">Klik Aku!</button>

<script>
  function tampilkanPesan() {
    alert("Tombol telah diklik!");
  }
</script>
```

**Penjelasan:**

* Fungsi `onclick` akan menjalankan kode saat tombol ditekan.
* Fungsi didefinisikan dalam `<script>`.

---

#### **3. Aktivitas: Buat Tombol Sendiri (25 menit)**

**Tugas siswa:**

* Buat tombol dengan label bebas (misal: “Cek nilai”, “Tebak umur”).
* Saat diklik, munculkan `prompt`, lalu tampilkan `alert` tergantung input.

**Contoh variasi:**

* Prompt nama → munculkan salam personal
* Prompt angka → jika angka > 10, beri komentar

---

### **C. Penutup (10 menit)**

**Guru menyimpulkan:**

> “Dengan `if` dan tombol, webmu bisa merespon pengguna secara dinamis. Di proyek berikutnya, kalian akan mulai membangun kalkulator interaktif sederhana!”

**Refleksi akhir:**

> “Apa perbedaan antara perintah `alert` dan `prompt`? Kapan kita pakai `if`?”

**Contoh tampilan dan interaksi langsung menggunakan editor online.**

#### **2. Penjelasan Struktur HTML Kalkulator (10 menit)**

**Contoh kode HTML:**

```html
<h2>Kalkulator Penjumlahan</h2>
<input type="number" id="angka1" placeholder="Angka pertama">
<input type="number" id="angka2" placeholder="Angka kedua">
<button onclick="jumlahkan()">Hitung</button>
<p>Hasil: <span id="hasil"></span></p>
```

**Penjelasan:**

#### **3. Penjelasan JavaScript Dasar Kalkulator (15 menit)**

```javascript
function jumlahkan() {
  let a = parseFloat(document.getElementById("angka1").value);
  let b = parseFloat(document.getElementById("angka2").value);
  let hasil = a + b;
  document.getElementById("hasil").innerText = hasil;
}
```

**Bahas:**

* Mengambil nilai dari input
* Mengubah string ke angka (`parseFloat`)
* Menampilkan hasil ke `<span>`

# **Bab 12: Proyek Akhir – Website Mini dengan Bantuan AI**

---

## **Struktur Umum Bab 12**

* **Durasi**: 2 pertemuan (masing-masing 80 menit)
* **Output**: Satu halaman web utuh berisi konten, desain, dan interaksi sederhana
* **Tools**: Replit, CodePen, atau editor HTML lokal + AI pendamping (ChatGPT/alternatif)

---

## **Pertemuan 1: Rancang dan Bangun Website Mini**

### **A. Pendahuluan (10 menit)**

**Guru membuka:**

> “Hari ini kalian akan mulai membangun **website minimu sendiri**. Boleh tentang diri sendiri, hobi, komunitas, atau topik favorit. Gunakan bantuan AI sebagai partner coding, bukan sebagai tukang copas. Tugas kalian adalah membuat web yang ‘hidup’ dan ‘bermakna’.”

**Tujuan disampaikan:**

* Siswa mampu merancang isi, tampilan, dan fungsi halaman.
* Siswa memanfaatkan AI sebagai alat bantu dalam menulis dan mengembangkan kode.

---

### **B. Isi Materi (60 menit)**

#### **1. Brainstorm & Perencanaan (15 menit)**

**Tugas siswa (boleh di kertas atau Google Docs):**

* Tema website (misal: diri sendiri, musik favorit, resep makanan)
* Struktur konten:

  * Judul halaman
  * Paragraf teks
  * Gambar
  * Interaksi (form, tombol, alert, dll)
* Gaya visual yang diinginkan (minimalis, warna cerah, tema gelap, dll)

---

#### **2. Membuat Website dengan Bantuan AI (45 menit)**

**Langkah:**

* Siswa menulis prompt ke AI, contoh:

> *“Buatkan halaman HTML berjudul 'Tentang Kucing', dengan paragraf deskripsi, satu gambar, tombol yang jika diklik akan menampilkan alert 'Meow!' dan warna latar biru.”*

**Proses:**

1. Salin dan uji kode di editor online.
2. Identifikasi bagian kode HTML, CSS, dan JavaScript.
3. **Modifikasi konten dan tampilan** agar sesuai dengan ide pribadi.

---

### **C. Penutup (10 menit)**

**Guru menyimpulkan:**

> “Hari ini kalian sudah mulai membangun website sendiri. Di pertemuan berikutnya, kita akan menyempurnakan dan mempresentasikan karya kalian.”

**Tugas rumah:**

> “Perbaiki halaman websitemu agar lebih menarik. Siapkan untuk ditampilkan minggu depan.”

---

## **Pertemuan 2: Penyempurnaan & Presentasi Website**

### **A. Pendahuluan (5–10 menit)**

**Guru membuka:**

> “Hari ini kita akan menyelesaikan proyek website mini kalian. Fokus utama adalah penyempurnaan dan presentasi—bagaimana kalian menyampaikan gagasan dengan visual dan interaksi.”

**Tujuan:**

* Menyempurnakan konten, tampilan, dan fungsi web.
* Melatih siswa mempresentasikan hasil karyanya.

---

### **B. Isi Materi (60 menit)**

#### **1. Tahap Penyempurnaan Mandiri (30 menit)**

**Siswa diberi waktu untuk:**

* Memperbaiki tata letak
* Memastikan interaksi berjalan baik
* Menambahkan elemen tambahan:

  * Ganti font
  * Tambahkan gambar latar
  * Tambahkan fitur `prompt` atau tombol tambahan

**Guru berperan sebagai pembimbing dan reviewer cepat.**

---

#### **2. Presentasi Mini Proyek (30 menit)**

**Format:**

* 1–2 menit per siswa/kelompok kecil
* Tampilkan halaman di layar atau share link
* Ceritakan:

  * Apa temanya?
  * Apa yang dibuat sendiri?
  * Apa yang dihasilkan oleh AI?
  * Apa bagian favorit dan yang paling menantang?

**Penilaian formatif oleh guru + apresiasi kelas.**

---

### **C. Penutup (10 menit)**

**Guru menyimpulkan dan refleksi akhir:**

> “Dari awal tahun kalian belajar membangun web—dari struktur, desain, logika, hingga cara memanfaatkan AI. Website mini kalian hari ini adalah bukti bahwa kalian bukan hanya pengguna teknologi, tapi pencipta.”

**Pertanyaan reflektif:**

* Apa yang paling kalian pelajari tentang proses membuat web?
* Bagaimana peran AI dalam membantumu belajar dan berkarya?
* Apakah kalian merasa siap membangun proyek yang lebih besar di masa depan?

---

## **Struktur Minimum Website Mini**

| Elemen Wajib        | Keterangan                                          |
| ------------------- | --------------------------------------------------- |
| Judul (`<h1>`)      | Judul halaman (topik bebas)                         |
| Paragraf (`<p>`)    | Deskripsi isi                                       |
| Gambar (`<img>`)    | Minimal 1 gambar relevan                            |
| Tombol (`<button>`) | Interaksi saat diklik (alert, prompt, ubah warna)   |
| CSS                 | Warna latar, font, border, tata letak               |
| JavaScript          | Minimal 1 fungsi interaktif (mengubah teks, prompt) |
