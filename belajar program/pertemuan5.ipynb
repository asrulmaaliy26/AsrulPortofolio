2. **Tipe Data Campuran**
   Jika ditulis:

   ```javascript
   let a = 10;
   let b = "2";
   console.log(a * b);
   ```

   Apa hasil output, dan mengapa tidak error padahal `b` adalah string?

---

3. **Operator Aritmatika**

4. **Debugging Error**
   Ada kode berikut:

   ```javascript
   let nama = "Naim";
   console.log("Halo, nama saya " nama);
   ```

   Mengapa kode di atas error? Bagaimana memperbaikinya?

---

5. **Boolean Expression**
   Diberikan kode:

   ```javascript
   let umur = 16;
   let izin = false;
   console.log(umur >= 17 && izin == true);
   ```

   Apa hasil yang ditampilkan? Jelaskan alasannya.

---

6. **Prediksi Output (typeof)**

   ```javascript
   let a = 12 / "3";
   console.log(a);
   console.log(typeof a);
   ```

   Apa hasil output kedua `console.log` tersebut?

---

7. **Variabel dan String Template**
   Gunakan *template literal* (backtick \`\`) untuk menampilkan kalimat:
   `Halo, nama saya Sinta dan umur saya 16 tahun`
   dengan variabel `nama = "Sinta"` dan `umur = 16`.

---

8. **Perbandingan**
   Apa perbedaan hasil dari kode berikut?

   ```javascript
   console.log(5 == "5");
   console.log(5 === "5");
   ```
